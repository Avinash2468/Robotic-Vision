```python
# Only numpy & matplotlib is sufficient for this question.
##############################################################################
# TODO: Do tasks described in Question 1                                     #
##############################################################################
# Replace "pass" statement with your code (You can split this cell into
# multiple cells if you wish to)

"""

Helper Functions

"""
def gaussiann(a,m,s,x):
    final = a*np.exp((-(x-m)**2)/(2*(s**2)))
    return final

def jacobiann(a,m,s,x_list):
    finalest = np.empty([1, 3])
    for x in x_list:
        first = np.exp((-(x-m)**2)/(2*(s**2)))
        second = ((a*(x-m))/s**2)*np.exp((-(x-m)**2)/(2*(s**2)))
        third = ((a*((x-m)**2))/s**3)*np.exp((-(x-m)**2)/(2*(s**2)))
        final = np.array([first,second,third])
        finalest= np.vstack((finalest,final))
    return finalest

def residuall(a,m,s,x,y):
    final = (a*np.exp((-(x-m)**2)/(2*(s**2)))) - y
    return final

"""

Hyperparameters

"""

#lr = 0.057
#lr = 0.01
#lr = 0.1
#lr = 0.1

num_iter = 200
tol = 1e-15
lamda = 0.01

"""

Code

"""
a = 10
m = 0
s = 20

#x_obs = np.linspace(-25,25,50)
#y_obs = gaussiann(a,m,s,x_obs)

#a_0 = 10
#m_0 = 13
#s_0 = 19.12


"""

Gradient Descent

"""
def find_gd(x_obs,num_iter,lr,ini,noise):

    tol = 1e-15
    a = 10
    m = 0
    s = 20
    
    pp =len(x_obs)
    y_obs = gaussiann(a,m,s,x_obs).reshape(pp,1)
    y_obs =y_obs + noise
    
    r_norm = []
    
    a_0 = ini[0]
    m_0 = ini[1]
    s_0 = ini[2]
    
    a_pred_gd = a_0 
    m_pred_gd = m_0
    s_pred_gd = s_0
    
    stop_iter = -1

    for j in range(num_iter):
        
        resi_final = []
        
        for i in range(len(x_obs)):
            resi = residuall(a_pred_gd,m_pred_gd,s_pred_gd,x_obs[i],y_obs[i])
            resi_final.append(resi)
        
        jayZ = jacobiann(a_pred_gd,m_pred_gd,s_pred_gd,x_obs)[1:,:]
        
        resi_final = (np.asarray(resi_final)).reshape(pp,1)
        
        update = -lr*jayZ.T@resi_final
        #update = -lr*update
        #print(update)
        a_pred_gd= a_pred_gd+update[0,0]
        m_pred_gd= m_pred_gd+update[1,0]
        s_pred_gd= s_pred_gd+update[2,0]
        
        r_norm.append(np.linalg.norm(resi_final)**2)
        #print(np.linalg.norm(update))
        #print(update)
        
        if( (abs(update[0,0])<tol) & (abs(update[1,0])<tol) & (abs(update[2,0])<2*tol) ):
            #print(np.linalg.norm(update))
            #print(update)
            stop_iter = j+1
            break
    
    if(stop_iter==-1):
        stop_iter = num_iter
    
    return a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm
            
        
        


"""

Gauss Newton

"""
def find_gn(x_obs,num_iter,lr,ini,noise):
    
    pp =len(x_obs)
    y_obs = gaussiann(a,m,s,x_obs).reshape(pp,1)
    y_obs += noise
    
    a_0 = ini[0]
    m_0 = ini[1]
    s_0 = ini[2]

    a_pred_gm = a_0 
    m_pred_gm = m_0 
    s_pred_gm = s_0     


    r_norm_gauss = []
    stop_iter = -1


    for j in range(num_iter):

        resi_final = []

        for i in range(len(x_obs)):
            resi = residuall(a_pred_gm,m_pred_gm,s_pred_gm,x_obs[i],y_obs[i])
            resi_final.append(resi)
        
        jayZ = jacobiann(a_pred_gm,m_pred_gm,s_pred_gm,x_obs)[1:,:]
        resi_final = (np.asarray(resi_final)).reshape(pp,1)

        update = np.linalg.pinv(jayZ.T@jayZ)@jayZ.T@resi_final
        a_pred_gm= a_pred_gm- update[0,0]
        m_pred_gm= m_pred_gm- update[1,0]
        s_pred_gm= s_pred_gm- update[2,0]
        r_norm_gauss.append(np.linalg.norm(resi_final)**2)
        
        if(np.linalg.norm(update)<tol):
            stop_iter = j+1
            break
            
    if(stop_iter==-1):
        stop_iter = num_iter
    
    return a_pred_gm,m_pred_gm,s_pred_gm,stop_iter,r_norm_gauss

    

"""

LM

"""
def find_lm(x_obs,num_iter,lamda,ini,noise,x_factor):
    
    a = 10
    m = 0
    s = 20
    tol = 1e-15
    
    pp =len(x_obs)
    y_obs = gaussiann(a,m,s,x_obs).reshape(pp,1)
    y_obs += noise
    
    a_0 = ini[0]
    m_0 = ini[1]
    s_0 = ini[2]

    ## ok
    a_pred_lm = a_0 
    m_pred_lm = m_0  
    s_pred_lm = s_0  

    r_norm_LM = []
    stop_iter = -1

    for j in range(num_iter):

        resi_final = []
        new_resi = []

        for i in range(len(x_obs)):
            resi = residuall(a_pred_lm,m_pred_lm,s_pred_lm,x_obs[i],y_obs[i])
            resi_final.append(resi)

        jayZ = jacobiann(a_pred_lm,m_pred_lm,s_pred_lm,x_obs)[1:,:]
        resi_final = (np.asarray(resi_final)).reshape(pp,1)

        n = jayZ.shape[0]
        j_h = jayZ.T@jayZ 
        I = np.identity(3)
        I[0,0] = j_h[0,0]
        I[1,1] = j_h[1,1]
        I[1,2] = j_h[2,2]
        

        update = np.linalg.pinv(j_h+ lamda*I)@jayZ.T@resi_final

        a_pred_lm=a_pred_lm - update[0,0]
        m_pred_lm=m_pred_lm - update[1,0]
        s_pred_lm=s_pred_lm - update[2,0]

        r_norm_LM.append(np.linalg.norm(resi_final)**2)
        
        for i in range(len(x_obs)):
            resi = residuall(a_pred_lm,m_pred_lm,s_pred_lm,x_obs[i],y_obs[i])
            new_resi.append(resi)
            
        new_resi = (np.asarray(new_resi)).reshape(pp,1)
        if (np.linalg.norm(new_resi)<np.linalg.norm(resi_final)):
            lamda = lamda/x_factor
        else:
            lamda = lamda*x_factor
            a_pred_lm=a_pred_lm + update[0,0]
            m_pred_lm=m_pred_lm + update[1,0]
            s_pred_lm=s_pred_lm + update[2,0]
            
            
        
        #print(tol)
        if(np.linalg.norm(update)<tol):
            #print(np.linalg.norm(update)<tol)
            stop_iter = j+1
            break
    
    if (stop_iter==-1):
        stop_iter = num_iter
            
    
    return a_pred_lm,m_pred_lm,s_pred_lm,stop_iter,r_norm_LM




##############
'''
plotting functions starts from here
'''
###############

def get_plot_gd(x_obs,num_iter,lr,ini,noise_mean,noise_var):
    
    pp = len(x_obs)
    print("Plots using gradient descent for num_iter=%f observations=%f  a_i = %f m_i = %f s_i = %f Learning_rate = %f Noise:mean=%f variance=%f"%(num_iter,pp,ini[0],ini[1],ini[2],lr,noise_mean,noise_var))
    
    pp = len(x_obs)
    noise = np.random.normal(noise_mean,noise_var,pp).reshape(-1,1)
    y_obs = gaussiann(a,m,s,x_obs)
    a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm =find_gd(x_obs,num_iter,lr,ini,noise)
    y_pred_gd = gaussiann(a_pred_gd,m_pred_gd,s_pred_gd,x_obs)
    r_norm_final = r_norm[-1]
    
    f, axarr = plt.subplots(1,2)
    f.set_figheight(7)
    f.set_figwidth(16)

    axarr[0].plot(x_obs,y_obs)
    axarr[0].plot(x_obs,y_pred_gd)
    
    axarr[0].set_title("Ground truth vs predicted values for Gradient Descent")

    axarr[1].plot(range(stop_iter),r_norm)
    axarr[1].set_title("Number of interations vs cost (||r^(2)||) for gradient descent")
    
    plt.show()
    #print(r_nor)
    print("The Value of the Cost Function before gradient descent is:" + str(r_norm[0]) + " at learning rate:%f "%(lr))
    print("The Value of the Cost Function after gradient descent is: %f  at learning rate:%f"%(r_norm[-1],lr))
    print("It went till " +str(num_iter)+ " interations and predicted outputs as a = %f m = %f s = %f" %(a_pred_gd,m_pred_gd,s_pred_gd,))
    print("\n")
    print("---------------------------------------------------------------------------")
    print("\n")
    
def get_plot_gn(x_obs,num_iter,ini,noise_mean,noise_var):
    
    lr=0
    pp = len(x_obs)
    
    print("Plots using Guass Newton for num_iter=%f observations=%f  a_i = %f m_i = %f s_i = %f  Noise:mean=%f variance=%f"%(num_iter,pp,ini[0],ini[1],ini[2],noise_mean,noise_var))

    noise = np.random.normal(noise_mean,noise_var,pp).reshape(-1,1)
    y_obs = gaussiann(a,m,s,x_obs)
    a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm =find_gn(x_obs,num_iter,lr, ini,noise)
    y_pred_gd = gaussiann(a_pred_gd,m_pred_gd,s_pred_gd,x_obs)
    r_norm_final = r_norm[-1]
    
    f, axarr = plt.subplots(1,2)
    f.set_figheight(7)
    f.set_figwidth(16)

    axarr[0].plot(x_obs,y_obs)
    axarr[0].plot(x_obs,y_pred_gd)
    axarr[0].set_title("Ground truth vs predicted values for Guass Newton")

    axarr[1].plot(range(stop_iter),r_norm)
    axarr[1].set_title("Number of interations vs cost (||r^(2)||) for Guass Newton")
    plt.show()
    
    print("The Value of the Cost Function before Guass Newton is:" + str(r_norm[0]) + " at learning rate:%f "%(lr))
    print("The Value of the Cost Function after Guass Newton is: %f  at learning rate:%f"%(r_norm[-1],lr))
    print("It went till " +str(num_iter)+ " interations and predicted outputs as a = %f m = %f s = %f" %(a_pred_gd,m_pred_gd,s_pred_gd,))
    print("\n")
    print("---------------------------------------------------------------------------")
    print("\n")

def get_plot_lm(x_obs,num_iter,lamda,ini,noise_mean,noise_var,x_factor):
    
    pp = len(x_obs)
    print("Plots using LM for num_iter=%f observations=%f  a_i = %f m_i = %f s_i = %f initial_lamda = %f Noise:mean=%f variance=%f"%(num_iter,pp,ini[0],ini[1],ini[2],lamda,noise_mean,noise_var))

    noise = np.random.normal(noise_mean,noise_var,pp).reshape(-1,1)
    y_obs = gaussiann(a,m,s,x_obs)
    a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm =find_lm(x_obs,num_iter,lr,ini,noise,x_factor)
    y_pred_gd = gaussiann(a_pred_gd,m_pred_gd,s_pred_gd,x_obs)
    r_norm_final = r_norm[-1]
    
    f, axarr = plt.subplots(1,2)
    f.set_figheight(7)
    f.set_figwidth(16)

    axarr[0].plot(x_obs,y_obs)
    axarr[0].plot(x_obs,y_pred_gd)
    axarr[0].set_title("Ground truth vs predicted values for LM")

    axarr[1].plot(range(stop_iter),r_norm)
    axarr[1].set_title("Number of interations vs cost (||r^(2)||) for LM")
    plt.show()
    
    
    print("The Value of the Cost Function before LM is:" + str(r_norm[0]) + " at learning rate:%f "%(lr))
    print("The Value of the Cost Function after LM is: %f  at learning rate:%f"%(r_norm[-1],lr))
    print("It went till " +str(num_iter)+ " interations and predicted outputs as a = %f m = %f s = %f" %(a_pred_gd,m_pred_gd,s_pred_gd,))
    print("\n")
    print("---------------------------------------------------------------------------")
    print("\n")

def lr_iteration(x_obs,num_iter,ini,noise,x_factor):
    
    lr = 0.001
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_gd(x_obs,num_iter,lr,ini,noise)
        lr_list.append(lr)
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        norm_final.append(r_norm[-1])
        lr = lr + 0.005
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs Learning rate for Gradient Descent')
    plt.xlabel('Lambda')
    plt.ylabel('Iterations')
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final,'r')
    plt.title('Final Cost function vs Learning rate for Gradient Descent')
    plt.xlabel('Lambda')
    plt.ylabel('Cost function')
    
    ####
    
    
    lr = 0.001
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_lm(x_obs,num_iter,lr,ini,noise,x_factor)
        lr_list.append(lr)
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        norm_final.append(r_norm[-1])
        lr = lr + 0.005
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs Lambda for LM')
    plt.xlabel('Lambda')
    plt.ylabel('Iterations')
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final,'r')
    plt.title('Final Cost function vs Lambda for LM')
    plt.xlabel('Lambda')
    plt.ylabel('Cost function')
    #lr_iteration()
    
    
def obs_iteration(lr,lamda,num_iter,ini,noise,x_factor):
    
    obv = 10
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,0,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_gd(x_obs,num_iter,lr,ini,noise)
        lr_list.append(obv)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        obv = obv + 10
        
    plt.figure(figsize=(16,7))
    
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs No.of.Observations for Gradient Descent')
    plt.xlabel('No of observations')
    plt.ylabel('Iterations')
    
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final, 'r')
    plt.title('Final costfunction values vs No.of.Observations for Gradient Descent')
    plt.xlabel('No of observations')
    plt.ylabel('Cost function')
    
    obv = 10    
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,0,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_gn(x_obs,num_iter,lr,ini,noise)
        lr_list.append(obv)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        obv = obv + 10
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs No.of.Observations for Guass Newton')
    plt.xlabel('No of observations')
    plt.ylabel('Iterations')
    
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final, 'r')
    plt.title('Final costfunction values vs No.of.Observations for Guass Newton')
    plt.xlabel('No of observations')
    plt.ylabel('Cost function')
    
    
    obv = 10
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,0,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_lm(x_obs,num_iter,lamda,ini,noise,x_factor)
        lr_list.append(obv)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        obv = obv + 10
        
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs No.of.Observations for LM')
    plt.xlabel('No of observations')
    plt.ylabel('Iterations')
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final,'r')
    plt.title('Final Cost function vs No.of.Observations for LM')
    plt.xlabel('No of observations')
    plt.ylabel('Cost function')
    
    
def var_ite(lr,lamda,num_iter,ini,noise,x_factor):
    
    obv = 50
    var = 1
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,var,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_gd(x_obs,num_iter,lr,ini,noise)
        lr_list.append(obv)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        var = var + 1
        
    plt.figure(figsize=(16,7))
    
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs No.of.Observations for Gradient Descent')
    plt.xlabel('Variance')
    plt.ylabel('Iterations')
    
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final, 'r')
    plt.title('Final costfunction values vs No.of.Observations for Gradient Descent')
    plt.xlabel('Variance')
    plt.ylabel('Cost function')
    
    #obv = 10    
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,var,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_gn(x_obs,num_iter,lr,ini,noise)
        lr_list.append(var)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        var = var + 1
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs No.of.Observations for Guass Newton')
    plt.xlabel('Variance')
    plt.ylabel('Iterations')
    
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final, 'r')
    plt.title('Final costfunction values vs No.of.Observations for Guass Newton')
    plt.xlabel('Variance')
    plt.ylabel('Cost function')
    
    
    #obv = 10
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,var,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_lm(x_obs,num_iter,lamda,ini,noise,x_factor)
        lr_list.append(var)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        var = var + 1
        
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs Variance for LM')
    plt.xlabel('Variance')
    plt.ylabel('Iterations')
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final,'r')
    plt.title('Final Cost function vs Variance for LM')
    plt.xlabel('Variance')
    plt.ylabel('Cost function')
    
    var = 10
    ite = [ ]
    lr_list = []
    norm_final = []
    for i in range(10):
        x_obs = np.linspace(-25,25,obv)
        noise = np.random.normal(0,0,obv).reshape(-1,1)
        a_pred_gd,m_pred_gd,s_pred_gd,stop_iter,r_norm = find_lm(x_obs,num_iter,lamda,ini,noise,var)
        lr_list.append(var)
        norm_final.append(r_norm[-1])
        if(stop_iter == num_iter):
            stop_iter = np.inf    
        ite.append(stop_iter)
        var = var + 10
        
        
    plt.figure(figsize=(16,7))
    plt.subplot(121)
    plt.plot(lr_list, ite, 'r')
    plt.title('Iterations vs Update factor for LM')
    plt.xlabel('Variance')
    plt.ylabel('Iterations')
    
    plt.subplot(122)
    plt.plot(lr_list, norm_final,'r')
    plt.title('Final Cost function vs Update factor for LM')
    plt.xlabel('Variance')
    plt.ylabel('Cost function')
    
        
        
    
pass
##############################################################################
#                             END OF YOUR CODE                               #

##############################################################################
```
