# DLT Explanation

## Finding the Matrix P
- DLT or Direct Linear Transform any object point X to image point x using the relation 

$$x_{3x1} =  P_{3x4}X_{4x1}$$

- Here the P matrix is the camera parameter matrix that includes both intrinsic and extrinsic camera parameters.
-  We need to estimate the 11 elements of P($\because$ the last element is anyway '1').
- Hence we need atleast '6' points to get a good estimate because each point gives two equations :

$$A^{T} = \left[\begin{array}{r}p_{11}&p_{12}&p_{13}&p_{14}\end{array}\right]$$
$$B^{T} = \left[\begin{array}{r}p_{21}&p_{22}&p_{23}&p_{24}\end{array}\right]$$
$$C^{T} = \left[\begin{array}{r}p_{31}&p_{32}&p_{33}&p_{34}\end{array}\right]$$

$$\therefore P = \left[\begin{array}{rrr}A^{T}\\B^{T}\\C^{T}\end{array}\right]$$


$$x_{3x1} = \left[\begin{array}{rr}x\\y\\1\end{array}\right]= \left[\begin{array}{rrr}A^{T}X/C^{T}X\\B^{T}X/C^{T}X\\1\end{array}\right]$$

- Writing the above matrix in equation form we get 
$$ -X^{T}A + xX^{T}C = 0$$
$$ -X^{T}B + yX^{T}C = 0$$
- The above equations are linear in variables $A$, $B$ and $C$, hence we can solve them simultaneously.

- Now we vectorize $P$ basically write a $3x4$ $P$ matrix as
$$ p_{12x1} = \left[\begin{array}{rrr}A\\B\\C\end{array}\right]$$

- Inorder to solve for $p$ 
$$a_{x}^{T} = \left[\begin{array}{r}-X^{T}_{1x4}&0_{1x4}&xX^{T}_{1x4}\end{array}\right]_{1x12}$$
$$a_{y}^{T} = \left[\begin{array}{r}0_{1x4}&-X^{T}_{1x4}&yX^{T}_{1x4}\end{array}\right]_{1x12}$$

- We know
$$a_{x}^{T}p = 0$$
$$a_{y}^{T}p = 0$$

- As discussed before we need atleast six point correspondences in 2D and 3D to solve for $p$, hence we stack all $a_{x_{i}}^{T}$ and $a_{y_{i}}^{T}$ in a Matrix $M$ as 

$$M_{2Ix12} = \left[\begin{array}{rrrrrrrr}a_{x_{1}}^{T}\\a_{y_{1}}^{T}\\......\\a_{x_{i}}^{T}\\a_{y_{i}}^{T}\\......\\a_{x_{I}}^{T}\\a_{y_{I}}^{T}\end{array}\right]$$

- And we have

$$Mp = 0 $$

- So we can say $p$ belongs to $nullSpace(M)$, which may be difficult to find directly but a good estimate would be using the SVD method as :

$$M = UDV^{T}$$

- The best approximation for $nullSpace(M)$ would be taking the eigen vector in $V$ corresponding to smallest eigen value, hence we take the last column of $V$ as our vectorised matrix $p$.

- Then we rearrange or reshape the vectorised matrix $p_{12x1}$ to $P_{3x4}$.

- We also need to normalise the matrix $P$ with its last element i.e.
$$P = P/p_{34}$$

- This is our required Matrix P.

## Finding K, R and T matrices from P

- After getting the $P$ matrix we can also find the Intrinsic camera parameters given by $K$, Extrinsic camera parameters given by Rotation matrix $R$ and vector $T$. 

$$P_{3x4} =  \left[\begin{array}{r} H_{3x3}| h_{3x1}\end{array}\right]$$

$$H = K_{3x3}R_{3x3}$$

$$h = -K_{3x3}R_{3x3}X_{0}$$

$$X_{0} = -H^{-1}h$$

- From $qr$ decomposition of $H^{-1}$ we get
$$H^{-1} = R^{T}K^{-1}$$

$$\text{Here, }q = R^{T} \text{ and } r = K^{-1}$$

$$\therefore R = q^{T} \text{, } K = r^{-1}\text{ and } T = -RX_{0} $$

- Homogenising $K$ we get 
$$K = K/k_{33}$$

$$\therefore P_{3x4} =  K_{3x3}\left[\begin{array}{r} R_{3x3}| T_{3x1}\end{array}\right]$$


```python
#DLT function takes as input 
#the image points and world points
#return a 3x4 matrix

def DLT(x,X):
    M = [];
    zeros = np.zeros(4);
    L = len(X);
    for i in range(L):
        M.append(np.hstack((-X[i],np.hstack((zeros,x[i][0]*X[i])))));
        M.append(np.hstack((zeros,np.hstack((-X[i],x[i][1]*X[i])))));
    M = np.array(M).reshape(2*L,12);
    U,D,VT= np.linalg.svd(M);
    P = np.array(VT[-1]).reshape(3,4);
    P = P/P[-1,-1]
    return P

P = DLT(x,X);
print("The Matrix P is")
print(P)
```

## Analysing P

The next part of this question is to analyse how good our estimate is. Report the reprojection error of your P matrix

```python
#Get reprojected points
def get_points(P,X):
    points = P@X.T
    points = points/points[2]
    points = points.T[:,0:2]
    return points

#Get reprojection error
def get_err(x, X, P):
    points = get_points(P, X)
    x = x[:,0:2];
    err = np.sum((x-points)**2)/2/len(x)
    return err
error = get_err(x, X, P)
print("The MSE between calculated and actual image points is",error)
```

Given the P matrix, we can estimate K, R and T. Write a function that returns K,R,T given P.
Note: 
P = K * [R | T] where K is a 3x3 matrix and [R | T], P are 3x4 matrices

```python
def ret_RKT(P):
    H = P[:,:3];
    h = P[:,-1];
    X0 = -np.linalg.inv(H)@h;
    q,r = np.linalg.qr(np.linalg.inv(H));
    R = q.T;
    T = -R@X0;
    K = np.linalg.inv(r);
    K = K/K[-1,-1];
    return R, K ,T
R,K,T = ret_RKT(P);
print("Rotation matrix R is")
print(R)
print("\nCamera matrix K is ")
print(K)
print("\nTranslation vector T is ")
print(T)
P_recovered = np.empty((3,4));
P_recovered[:3,:3] = K@R;
P_recovered[:,-1] = K@T;
P_recovered /= P_recovered[-1,-1]
print("\nThe recovered P matrix from R, K, T and actual P matrix is same :",np.linalg.norm(P-P_recovered)<1e-5)
```

### Zhangs method

For this task, use the inbuilt Opencv function to estimate the K matrix of a camera. Use the checkerboard images 5456-5470 in images/zhangs folder for this task. Familiarize yourself with the different parameters for this method.

```python
#Code for zhangs here

# termination criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

#get images
images=[]
for i in range(56,71):
    fname ="./Images/Zhangs/IMG_54"+str(i)+".JPG"
    images.append(fname)

# preparing object points
objp = np.zeros((8*6,3), np.float32)
objp[:,:2] = np.mgrid[0:8,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = [] # 3d point in real world space
imgpoints = [] # 2d points in image plane.


for fname in images:
    img = cv2.imread(fname)
    
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chess board corners
    ret, corners = cv2.findChessboardCorners(gray, (8,6),None)

    # If found, add object points, image points (after refining them)
    if ret == True:
        
        objpoints.append(objp)
        corners2 = cv2.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
        imgpoints.append(corners2)

        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (8,6), corners2,ret)
        print("The corners in the image " + fname +" are ")
        plt.imshow(img)
        plt.show()
        cv2.waitKey(500)
cv2.destroyAllWindows();
```

Draw epipolar lines on the first image for the corresponding points in the second image.

```python
data_file = "q2/data.txt"

with open(data_file, 'r') as f:
    A = f.readlines()
    
img_points1 = A[2:12]
img_points1 = [i.strip()[:-1] for i in img_points1]
img_points1[0] = img_points1[0][1:]
img_points2 = A[15:25]
img_points2 = [i.strip()[:-1] for i in img_points2]
img_points2[0] = img_points2[0][1:]

F_mat = A[28:31]
F_mat = [i.strip()[:-1] for i in F_mat]
F_mat[0] = F_mat[0][1:]

img_1 = []
img_2 = []
F_MAT = []

for i in img_points1:
    img_1.append(np.fromstring(i[1:-1], sep=' '))
    
for i in img_points2:
    img_2.append(np.fromstring(i[1:-1], sep=' '))
    
for i in F_mat:
    F_MAT.append(np.fromstring(i[1:-1], sep=' '))

img_1 = np.array(img_1,int)
img_2 = np.array(img_2,int)
F_mat = np.array(F_MAT)
```

#  Finding the Epipolar Lines 

## Using the Coplanarity Equation

- F is the fundamental matrix and x and x' are the locations of the corresponding points in the first and second image respectively. The coplanarity constraint says that, $x'^{T}Fx = 0$.

## For image 1

- Homogenize the corresponding points x by adding a column of ones to them.
- Epipolar lines using points x in first image is given by $Fx = l_{1}, (\because l_{1}^Tx' = 0$ from Coplanarity Equation), where $l_{1}$ is the epipolar line corresponding to the points in first image.

## For image 2

- Homogenize the corresponding points x' by adding a column of ones to them
- Epipolar lines using points x' in second image is given by $F^{T}x' = l_{2}, (\because l_{2}^{T}x = 0$ from Coplanarity Equation), where $l_{2}$ is the epipolar line corresponding to the points in second image.  


## Drawing and Plotting Epipolar lines
- Using the implemented function $drawlines(img,lines,pts)$ we draw the epipolar lines obtained above using the corresponding points on opposite images i.e. $l_{2}$ is plotted on image 1 and $l_{1}$ is plotted on image 2.
- $plt.imshow$ is used to finally plot the images with the epipolar lines drawn on them.

```python
#Write code to plot epipolar lines on the images
img1 = plt.imread("q2/img1.jpg")
img2 = plt.imread("q2/img2.jpg")

#Homgenising the img points
img_1_final = np.append(img_1,np.ones((len(img_1),1)), axis=1)
img_2_final = np.append(img_2,np.ones((len(img_2),1)), axis=1)

#Finding the epipolar lines for img2 and img1 respectively
lines_1 = np.dot(F_mat, img_1_final.T).T
lines_2 = np.dot(F_mat.T,img_2_final.T).T

#Function to plot the epipolar lines on the images
def drawlines(img,lines,pts):
    np.random.seed(5)
    r,c,_ = img.shape
    for r, pt1 in zip(lines, pts):
        color = tuple(np.random.randint(0,255,3).tolist())
        x0,y0 = map(int, [0, -r[2]/r[1] ])
        x1,y1 = map(int, [c, -(r[2]+r[0]*c)/r[1]])
        img = cv2.line(img, (x0,y0), (x1,y1),color,5)
        img = cv2.circle(img, tuple(pt1),8,color,10)
    return img

#Drawing the epipolar lines on image 1
img1_epipoles = drawlines(img1,lines_2,img_1);
plt.imshow(img1_epipoles,aspect = 'auto')
plt.title('Epipolar Lines with points corresponding to Image 2 on Image 1',fontweight ="bold")
plt.show()

#Drawing the epipolar lines on image 2
img2_epipoles = drawlines(img2,lines_1,img_2);
plt.imshow(img2_epipoles,aspect = 'auto')
plt.title('Epipolar Lines with points corresponding to Image 1 on Image 2',fontweight ="bold") 
plt.show()
```

#  Finding the Epipolar Points

- All the epipolar lines will converge at the epipole. If we want to try to find the coordinates of the epipole without using the epipolar lines, we can make a good estimate from the fundamental matrix itself.

## For image 1
- We can use the fact that the equation of epipolar lines to points corresponding in first image is $Fx = l_{1}$, hence epipole should actually belong to $nullspace(F)$, ($\because Fe = 0$ when $e$ is the epipolar point) but a good estimation would be doing SVD i.e. $U, D, V^{T}$ and finding the eigen vector of $V$ corresponding to smallest eigen value. 

## For image 2
- We can use the fact that the equation of epipolar lines to points corresponding in second image is $F^{T}x' = l_{2}$, hence epipole should actually belong to $nullspace(F^{T})$,  
($\because F^{T}e' = 0$ when $e'$ is the epipolar point) but a good estimation would be doing SVD i.e. $U, D, V^{T}$ and finding the eigen vector of $V$ corresponding to smallest eigen value.


## Getting the Pixel location from Eigen vector
- In this case the eigen vector will be of length '3' so we divide the eigen vector with its last element to get the epipole location in the image by taking only the first 2 coordinates.
- Since we can get a decimal value, we round off the coordinates to nearest integer using $float$ because Pixels should be integer.

```python
u, d, v = np.linalg.svd(F_mat, full_matrices=False)
e = v[-1];
e = e/e[2]
e = np.floor(e[0:2])
print("Epipole of first image is at pixel location : (",e[0],",",e[1],")");

u, d, v = np.linalg.svd(F_mat.transpose(), full_matrices=False)
e = v[-1];
e = e/e[2]
e = np.floor(e[0:2])
print("Epipole of second image is at pixel location : (",e[0],",",e[1],")");
```
